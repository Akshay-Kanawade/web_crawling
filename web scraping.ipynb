# flipkart mobile info scrap

```python
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get('https://www.flipkart.com/search?p%5B%5D=facets.brand%255B%255D%3DSamsung&sid=tyy%2F4io&sort=recency_desc&wid=1.productCard.PMU_V2_1')
req
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
soup.prettify()
```

```python
data=soup.findAll('div',{'class':'_13oc-S'})
data
```

```python
title=soup.findAll('div',{'class':'_4rR01T'})
list2=[]
for t in title:
    list2.append(t.text)
#print(len(list2))    
#print(list2)
```

```python
info=soup.findAll('div',{'class':'fMghEO'})
list1=[]
for inf in info:
    inf1=inf.findAll('li',{'class':'rgWa7D'})
    list3=[]
    for item in inf1:
        list3.append(item.text)
    list1.append(list3)
#print(len(list1))
#print(list1)
```

```python
mobile_info=[]

for j in list2:
    dict1={}
    dict1['Title']=j
    for l in list1:
        dict1['Ram/Rom']=l[0]
        dict1['Display']=l[1]
        dict1['Camera']=l[2]
        dict1['Battery']=l[3]
        dict1['Warranty']=l[4]
    mobile_info.append(dict1)
```

```python
#print(mobile_info[0])
#print(mobile_info)
#print(len(mobile_info))
print("--------------------------------------------------------------------------------------")
for i in range(len(mobile_info)):
    print()
    print("Titel : ",mobile_info[i]['Title'])
    print("Ram/Rom : ",mobile_info[i]['Ram/Rom'])
    print("Display : ",mobile_info[i]['Display'])
    print("Camera : ",mobile_info[i]['Camera'])
    print("Battery : ",mobile_info[i]['Battery'])
    print("Warranty : ",mobile_info[i]['Warranty'])
    print("-------------------------------------------------------------------------------------")
```
